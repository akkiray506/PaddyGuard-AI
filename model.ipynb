```python
import os, kagglehub

path = kagglehub.dataset_download("loki4514/rice-leaf-diseases-detection")
print("Base path:", path)
print("Top folders:", os.listdir(path))
```

```python
import os

data_path = os.path.join(path, "Rice_Leaf_Diease", "Rice_Leaf_Diease")

train_path = os.path.join(data_path, "train")
test_path  = os.path.join(data_path, "test")

print("Train:", os.listdir(train_path))
print("Test:", os.listdir(test_path))
```

```python
import os
#Remove spaces
#Keep naming format consistent
#Avoid errors during dataset loading
rename_map = {
    "Sheath Blight": "sheath_blight",
    "Tungro": "tungro",
    "Neck_Blast": "neck_blast",
    "Rice Hispa": "rice_hispa",
}

for old, new in rename_map.items():
    old_path = os.path.join(test_path, old)
    new_path = os.path.join(test_path, new)
    if os.path.exists(old_path) and not os.path.exists(new_path):
        os.rename(old_path, new_path)

print("✅ Fixed Test:", os.listdir(test_path))
```

```python
import tensorflow as tf
#bulid deep learning models
#load image datasets
#train and evaluate the model

IMG_SIZE = (224, 224) #All images are resized to this size so the model receives uniform input.
BATCH_SIZE = 16#group size of 16
SEED = 42

#This loads images from the train folder.

train_ds = tf.keras.utils.image_dataset_from_directory(
    train_path,
    validation_split=0.2,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)
#split-80:20

val_ds = tf.keras.utils.image_dataset_from_directory(
    train_path,
    validation_split=0.2,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)
#this creates validation dataset

test_ds = tf.keras.utils.image_dataset_from_directory(
    test_path,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

class_names = train_ds.class_names
print("✅ Classes:", class_names)

#This code loads the rice leaf disease images from the dataset
#folders, resizes them to 224×224, splits the data into
#training and validation sets, prepares batches for model
#training, and extracts the class labels automatically from
#folder names.
```

```python
print("✅ Fixed Test:", os.listdir(test_path))
```

```python
import tensorflow as tf

IMG_SIZE = (224, 224)
BATCH_SIZE = 16
SEED = 42

train_ds = tf.keras.utils.image_dataset_from_directory(
    train_path,
    validation_split=0.2,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    train_path,
    validation_split=0.2,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    test_path,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

class_names = train_ds.class_names
num_classes = len(class_names)
print("✅ Classes:", class_names)

AUTOTUNE = tf.data.AUTOTUNE#This allows TensorFlow to automatically optimize dataset loading
train_ds = train_ds.shuffle(1000, seed=SEED).prefetch(AUTOTUNE)
val_ds   = val_ds.prefetch(AUTOTUNE)
test_ds  = test_ds.prefetch(AUTOTUNE)
#Dataset shuffling and prefetching techniques are applied to improve training performance and efficiency.
```

```python
from tensorflow.keras import mixed_precision
#This module allows TensorFlow to use different numeric data types during training.
mixed_precision.set_global_policy("mixed_float16")
print("✅ Policy:", mixed_precision.global_policy())
#This step enables mixed precision training, where the model uses float16 and float32 data types together to improve training speed and reduce memory usage
```

```python
from tensorflow.keras import layers
#This module provides different neural network layers such as:

#Pooling layers
#Dropout
#Dense layers
#Data augmentation layers.

data_aug = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.08),
    layers.RandomZoom(0.10),
    layers.RandomContrast(0.10),
], name="data_augmentation")

#Here you are using Transfer Learning.
#EfficientNetB0 is loaded
#Pretrained on ImageNet dataset
#The top classification layer is removed (include_top=False).
#Pretrained on ImageNet dataset
base = tf.keras.applications.EfficientNetB0(
    include_top=False,
    weights="imagenet",
    input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3)
)

#Feature Extraction Stage.

base.trainable = False  # Stage 1

#Define Model Input


inputs = tf.keras.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))

#Apply Data Augmentation

#The input images are randomly modified during training.

#This improves model robustness.

x = data_aug(inputs)
x = tf.keras.applications.efficientnet.preprocess_input(x)
x = base(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.3)(x)

# dtype float32 is important when using mixed precision
outputs = layers.Dense(num_classes, activation="softmax", dtype="float32")(x)

model = tf.keras.Model(inputs, outputs)
model.summary()


#A pretrained EfficientNetB0 model is used for feature extraction, and custom layers such as global average pooling, dropout, and a softmax classification layer are added to classify rice leaf diseases.
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=["accuracy"]
)

callbacks = [
    tf.keras.callbacks.ModelCheckpoint(
        "best_model.keras", monitor="val_accuracy", save_best_only=True, verbose=1
    ),
    tf.keras.callbacks.EarlyStopping(
        monitor="val_accuracy", patience=4, restore_best_weights=True, verbose=1
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss", factor=0.2, patience=2, min_lr=1e-6, verbose=1
    )
]
```

```python
history1 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=12,
    callbacks=callbacks
)
```

```python
model = tf.keras.models.load_model("best_model.keras")

base = model.get_layer("efficientnetb0")
base.trainable = True

# Unfreeze last 30 layers only
for layer in base.layers[:-30]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=["accuracy"]
)

history2 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=8,
    callbacks=callbacks
)
```

```python
model = tf.keras.models.load_model("best_model.keras")
test_loss, test_acc = model.evaluate(test_ds, verbose=1)
print("✅ Test Accuracy:", test_acc)
```
